```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from geopy.geocoders import Nominatim
import re
```

```python
pd.set_option("display.max_columns", None) # set_option 파마미터에 None값을 전달하면 모든 columns를 다 볼 수 있어요. (default= 20cols)
df = pd.read_csv('Kia_usedCar_0528(uts8).csv') # 전처리 전 raw data
df.head(10)
```

```python
len(df)
```

```python
# 잦은 크롤링으로 인해 중복값이 엄청 많습니다. car_url을 기준으로 제거해주겠습니다
df = df.drop_duplicates(subset='car_url')
len(df) # 44342 , 약 15,000개 행이 제거되었습니다.
```

```python
df.isnull().sum().sum() # 99
df=df.dropna(axis=0)
df.isnull().sum().sum() # 0

# 공백값이 있는 행들을 제거 해주었기 때문에 DataFrame의 index가 비어있습니다.
# reset_index를 통해 정수형 인덱스를 다시 부여해줍니다.
df = df.reset_index(drop=True)
```

```python
# 사용개월 수(use) columns를 변환해주는 함수입니다.
def get_months(use_months):  # 'Y년 M개월'의 형태를 가져야 함(str)
    
    try:
        x = int(use_months.split()[0].replace('년','')) * 12
        y = int(use_months.split()[1].replace('개월',''))
        
    except Exception as e:
        
        if '년' in use_months:
            x = int(use_months.replace('년','')) *12
            y = 0
        elif '개월' in use_months:
            x = 0
            y = int(use_months.replace('개월',''))
        else:
            print(e)
    
    months = x+y
    
    return months

# 연식(year) columns를 변환해주는 함수입니다.
def change_year(car_year): #'YY년 MM개월(YY년식)'의 형태를 가져야 함(str)
    car_year = car_year.split('(')[-1].replace(')','')
    if car_year[0] == '9':
        car_year = int('19'+car_year.replace('년식',''))
    else:
        car_year = int('20'+car_year.replace('년식',''))
    
    return car_year

# folium을 후에 구현하기 위해 판매업체 주소(sales_loca) columns를 변환해주는 함수입니다.
def check_sales_loca(area_text,loca_text): # 한국 주소지의 형태를 가져야함 str
    
    if area_text == loca_text[:2]:
        loca_text = loca_text.split()[0]+' '+loca_text.split()[1]
        
    else :
        loca_text = area_text
        
    return loca_text

# 상위 3개의 함수와 다른 전처리를 통해서 raw_dataframe을 가공해주는 함수입니다.
def change_data(dataframe):
   
    for i in range(len(dataframe)):
        try:
            dataframe['car_name'][i] = dataframe['car_name'][i].replace(dataframe['car_brand'][i]+' ' ,'')
            dataframe['price'][i] = dataframe['price'][i].replace('만원','')
            dataframe['depreciation'][i] = dataframe['depreciation'][i].replace('%','')
            dataframe['mileage'][i] = dataframe['mileage'][i].replace('km','')
            dataframe['forecast_max'][i] = dataframe['forecast_max'][i].replace('만원','')
            dataframe['car_cc'][i] = dataframe['car_cc'][i].replace('cc','')
            dataframe['use'][i] = get_months(use_months=dataframe['use'][i])
            dataframe['year'][i] = change_year(car_year=dataframe['year'][i])
            dataframe['sales_loca'][i] = check_sales_loca(area_text=dataframe['car_area'][i],loca_text=dataframe['sales_loca'][i])
        except Exception as e:
            print(f'{i}번째 {e}에러')
    
    changed_df = dataframe
        
    
    return changed_df
```

```python
df2 = change_data(df) #위에 만들어놓은 함수에 raw_dataframe을 넣어줍니다.
df2 = df2.astype({'car_url':'int','price':'int', 'use':'int', 'depreciation':'int', 'mileage':'int',
                  'forecast_min':'int', 'forecast_max':'int', 'car_cc':'int'}) # DataFrame의 값의 형변환은 .astype()을 사용합니다.
```

```python
def get_new_price(price,depreciation):
    new_price = int((price/(100 - depreciation)) * 100)
    
    return new_price
```

```python
df2.insert(6,'new_price',0) # df에서 price columns앞에 위치시키겠습니다.(값은 0으로 채워서 int타입으로 만들어줍니다)
for i in range(len(df2)):
    df2['new_price'][i] = get_new_price(df2['price'][i], df2['depreciation'][i]) # for문을 이용해서 함수를 적용해줍니다.

df2.head()
```

```python
df2.info() # 모든 값의 형변환도 잘 되었는지 확인합니다.
```

```python
def get_lat_lng(address, user_agent = 'South Korea'):
    
    try:
        geo_local = Nominatim(user_agent = user_agent)
        geo = geo_local.geocode(address)
        lat, lng = geo.latitude, geo.longitude
    
    except AttributeError as e:
        lat, lng = 37.60, 127.00 # 대체주소 위도, 경도
        print("AttributeError: ", e, "대체주소({lat},{lng})로 반환합니다")
    
    return lat, lng
```

```python
df2.insert(26,'lat',0.0)
df2.insert(27,'lng',0.0)
```

```python
loca_list = list(df2['sales_loca'].unique()) # loca_list에 주소 고유값들이 들어있는 list가 만들어졌습니다.

loca_dict = {}

for loca in tqdm(loca_list): # loca_list에서 loca를 하나씩 받아서
    
    lat,lng = get_lat_lng(address =loca) # 위도,경도를 구해주고
    
    loca_dict[f'{loca}'] = [lat,lng] # loca_dict{ key: loca , value : [위도, 경도] }를 추가해줍니다.
```

```python
for i,df2_loca in enumerate(tqdm(df2['sales_loca'])):
    
    df2['lat'][i] = loca_dict.get(df2_loca)[0]
    df2['lng'][i] = loca_dict.get(df2_loca)[1]
```

```python
df['options_count'] = 0

# 문자열을 받아서 전처리 후 리스트로 변환해줍니다.
# 주의점 !!!: 능력이 부족하여, df의 options가 아무런 처리가 되어 있지 않은 상태의 str타입이어야 합니다
# for문 중간에 멈춰서 다시 돌리면, 앞서 변환되었던 df['option']의 값이 type:list라서 오류납니다.ㅎㅎㅎㅎ...;;

def options_to_word_list(option_text,stopwords = ['등록된','주요','옵션이','없습니다']):
    
    try:
        option_text = re.sub('[^ㄱ-ㅎ가-힣0-9\\s]','',option_text)
        word_list = option_text.split()
        word_list = [word for word in word_list if word not in stopwords]
    except:
        word_list = []
    
    return word_list
```

```python
for i,option in enumerate(tqdm(df['options'])):
        df['options'][i] = options_to_word_list(option)
df.head() # [등록된/옵션이/없습니다.]는 []로 반환되고, 나머지는 리스트형태로 반환됩니다.
```

```python
for i,options in enumerate(tqdm(df['options'])):
    df['options_count'][i] = len(options)

df.head()
```
