## Categorical Data 분포 확인 

```python
c_feature = ['car_area','car_brand','car_model','fuel','car_type','insurance']
```

```python
df_c = df[c_feature]
df_c
```

```python
freq_list = []
label_list = []
for col in df_c.columns:
    find_mode = col
    mode_uniq_val = df_c[find_mode].unique()
    val_freq = []

    for uniq_val in mode_uniq_val:
        val_freq.append(len(df_c[df_c[find_mode]== uniq_val]))

    #print(mode_uniq_val)
    #print(val_freq)
    
    frequency = sorted(zip(mode_uniq_val,val_freq),key=lambda x: x[1],reverse=True)
    #print(frequency)
    
    sorted_uniq_val = [x[0] for x in frequency[:10]]
    sorted_val_freq = [x[1] for x in frequency[:10]]
    
    freq_list.append(sorted_val_freq)
    label_list.append(sorted_uniq_val)
    #print(sorted_val_freq)
    #print(sum(sorted_val_freq))
    
print(freq_list)
print(label_list)
```

## 상위 10개만 표시

```python
plt.figure(figsize=(40,20))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.3)
colors = ['#ffc000','#ff9999','#ff4500','#ffa500','#ffe4b5', '#f4a560', '#f5deb3','#b0e0e6','#800000','#FFFF00']


for i in range(len(label_list)):
    plt.subplot(2,3,i+1)
    plt.title(df_c.columns[i].upper(), fontsize = 20)
    explode = [0.05 for x in range(len(freq_list[i]))]
    plt.pie(freq_list[i],
            labels = label_list[i],
            autopct='%.1f%%',
           startangle = 90,
           counterclock=False,
           textprops = {'fontsize':20},explode=explode, shadow=True, colors = colors)    
```

```python
fig = plt.figure(figsize=(15,10))
colors = ['#ffc000','#ff9999','#ff4500','#ffa500','#ffe4b5', '#f4a560', '#f5deb3','#b0e0e6','#800000','#FFFF00']

plt.title('중고차 매물 비율', fontsize = 18)
explode = [0.05 for x in range(10)]
plt.pie(freq_list[1],
        labels = label_list[1],
        autopct='%.1f%%',
       startangle = 90,
       counterclock=False,
       textprops = {'fontsize':15},explode=explode, colors= colors)
plt.show()
```

```python
total_youtube = pd.read_csv('labeled_youtube_0605.csv', lineterminator='\n')
related_df = pd.read_csv('preprocessed_emotion_labeled_youtube_0607.csv', lineterminator='\n')
```

```python
related_df.dropna(inplace=True)
related_df.reset_index(inplace=True)
related_df.info()
```

```python
fig = plt.figure(figsize=(15,10))

title = ['Youtube 영상 댓글 비율',
         '자동차 관련 댓글 감성 비율']

data = [[len(total_youtube[total_youtube['related'] == 1]), len(total_youtube[total_youtube['related'] == 0])],
        [len(related_df[related_df['emotion'] == 0]), len(related_df[related_df['emotion'] == 1]),len(related_df[related_df['emotion'] == -1])]]

labels = [['관련', '비관련'],['긍정','부정','중립']]

plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.3)

for i in range(0,2):
    plt.subplot(1,2,i+1)
    plt.title(title[i], fontsize = 20)
    explode = [0.05 for x in range(len(labels[i]))]
    plt.pie(data[i],
            labels = labels[i],
            autopct='%.1f%%',
           startangle = 90,
           counterclock=False,
           textprops = {'fontsize':18}, shadow=True, explode=explode, colors=['#ffc000','#ff9999','#ff4500'])

#plt.text(0,0,'s')

plt.show()
```

```python
brand_list = ['그랜저', '모닝', '카니발', '포터', '쏘나타', 'K5', '아반떼', '쏘렌토', '스파크', '봉고']
freq_list = []

for i in range(len(brand_list)):
        freq = len(related_df[related_df['comments'].str.contains(brand_list[i])])
        freq_list.append(freq)
    
freq_list
```

```python
frequency = sorted(zip(brand_list,freq_list),key=lambda x: x[1],reverse=True)
#print(frequency)

sorted_uniq_val = [x[0] for x in frequency[:10]]
sorted_val_freq = [x[1] for x in frequency[:10]]
```

```python
fig = plt.figure(figsize=(15,10))
colors = ['#ffc000','#ff9999','#ff4500','#ffa500','#ffe4b5', '#f4a560', '#f5deb3','#b0e0e6','#800000','#FFFF00']

plt.title('유튜브 댓글  언급 비율', fontsize = 18)
explode = [0.05 for x in range(len(sorted_uniq_val))]
plt.pie(sorted_val_freq,
        labels = sorted_uniq_val,
        autopct='%.1f%%',
       startangle = 90,
       counterclock=False,
       textprops = {'fontsize':15},explode=explode, colors= colors)
plt.show()
```
