```python
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}
driver = webdriver.Chrome("./chromedriver.exe")

try: 
    url = 'https://www.car365.go.kr/web/contents/websold_vehicle.do'
    res = requests.get(url, headers = headers)
    res.raise_for_status()

except Exception as e:
    print(e)
    
else:
    driver.get(url)
    driver.maximize_window()
    driver.implicitly_wait(10)
    
    for car_no in tqdm(df['car_no'][:4]):
        
        search = driver.find_element_by_xpath('//*[@id="search_str"]')
        search.click()
        search.send_keys(car_no) #네이버 검색창에 네이버 항공권 입력
        search.send_keys(Keys.ENTER)
        sleep(2)
        try:
            mile = driver.find_element_by_xpath('//*[@id="GUILD_CAR_MILEAGE"]').text
        except:
            print(f"{car_no}미등록")
            pass
        else:
            print(mile)
            search.click()
            search.send_keys(Keys.CONTROL + "a")
            search.send_keys(Keys.DELETE)
```
