```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tqdm import tqdm

pd.set_option("display.max_columns", None)
```

```python
os.listdir()
```

```python
file_list = os.listdir()
```

```python
len(file_list)
```

```python
file_list[2].endswith('.csv')
```

```python
youtube_csv_list = [csv for csv in file_list if csv.endswith('.csv')]
```

```python
youtube_csv_list[22]
```

```python
for youtube_csv in tqdm(youtube_csv_list):
    
    brand = youtube_csv.replace('.csv','').split('+')[0]
    model = youtube_csv.replace('.csv','').split('+')[1]
    
    if brand == model :
        pass
   
    else :

        try:
            df = pd.read_csv(youtube_csv, lineterminator='\n')
            df = df[['comments','likes']]
            df.insert(0,'car_brand',brand)
            df.insert(1,'car_model',model)
            df.to_csv('Pre+'+youtube_csv, index = False)

        except:
            print(youtube_csv)
            pass
```

```python
df.info()
```

```python
df_list = []

for idx,youtube_csv in enumerate(tqdm(youtube_csv_list)):
    
    df = pd.read_csv(youtube_csv,lineterminator='\n')
    df.rename(columns = {'likes\r':'likes'}, inplace = True)
    df_list.append(df)
```
