### 성능조회, 보험이력과 가격의 상관관계

```python
from sklearn import preprocessing
```

```python
# 벤츠 E클래스 (W213) 행만 뽑기
benz_eclass = car_information[car_information['car_name'] == '벤츠 E클래스 (W213)']
```

```python
benz_eclass.shape
```

```python
# 원-핫 인코딩

def encode_features(dataDF):
    features  = ['loss','flood','usage','change','insurance']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF
    
car_information = encode_features(benz_eclass)
car_information
```

```python
plt.rc("font", family = 'AppleGothic')
plt.subplots(figsize=(10,8))
sns.heatmap(benz_eclass[['price','loss','flood', 'usage', 'change', 'insurance']].corr(),
           annot=True)
```
