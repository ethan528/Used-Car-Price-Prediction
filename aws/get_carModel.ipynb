### 댓글에 없는 브랜드가 많아서 우선 댓글 기준으로 모델명 정리
---

```python
brand_list = youtube_df['car_brand'].unique()
model_list = youtube_df['car_model'].unique()
```

```python
for brand in tqdm(brand_list):
    search_brand = brand
    for model in model_list:
        search_model = model
        
        model_idx_list = df[(df['car_brand'].str.contains(search_brand))&(df['car_name'].str.contains(search_model))].index

        if len(model_idx_list) > 0:
            for idx in model_idx_list:
                df['car_model'][idx] = search_model

df[df['car_name'].str.contains(search_model)]
```

```python
# 중고차 DataFrame에서 'car_model' 빈 부분 찾기

df[df['car_model'] =='']['car_name'].unique()
```

## 1. 한국GM에 모델명에 '쉐보레 '삭제해주고 다시 돌리기


```python
youtube_df[youtube_df['car_brand'].str.contains('한국GM')]['car_model'].unique()
```

```python
search_brand = '한국GM'
model_list2 = youtube_df[youtube_df['car_brand'].str.contains(search_brand)]['car_model'].unique()

for model in model_list2:
    search_model = model.replace('쉐보레 ','')

    model_idx_list = df[(df['car_brand'].str.contains(search_brand))&(df['car_name'].str.contains(search_model))].index

    if len(model_idx_list) > 0:
        for idx in model_idx_list:
            df['car_model'][idx] = search_model

df[df['car_name'].str.contains(search_model)]
```

```python
# 중고차 DataFrame에서 'car_model' 빈 부분 찾기

df[df['car_model'] =='']['car_name'].unique()
```

## 2. 르노삼성 >> 르노코리아 처리해주기


```python
youtube_df[youtube_df['car_brand'].str.contains('르노')]
```

```python
# 르노코리아 = 르노삼성

search_brand = '르노'
model_list2 = youtube_df[youtube_df['car_brand'].str.contains(search_brand)]['car_model'].unique()

for model in model_list2:
    search_model = model.replace('쉐보레 ','')

    model_idx_list = df[(df['car_brand'].str.contains(search_brand))&(df['car_name'].str.contains(search_model))].index

    if len(model_idx_list) > 0:
        for idx in model_idx_list:
            df['car_model'][idx] = search_model

df[df['car_name'].str.contains(search_model)]
```

```python
# 중고차 DataFrame에서 'car_model' 빈 부분 찾기

df[df['car_model'] =='']['car_name'].unique()
```

## 3. 벤츠 S-클래스 >> 벤츠 S클래스 처리해주기

```python
# 클래스 중간에 하이픈 삭제

search_brand = '벤츠'
model_list2 = youtube_df[youtube_df['car_brand'].str.contains(search_brand)]['car_model'].unique()

for model in model_list2:
    search_model = model.replace('-','')

    model_idx_list = df[(df['car_brand'].str.contains(search_brand))&(df['car_name'].str.contains(search_model))].index

    if len(model_idx_list) > 0:
        for idx in model_idx_list:
            df['car_model'][idx] = search_model

df[df['car_name'].str.contains(search_model)]
```

```python
# 중고차 DataFrame에서 'car_model' 빈 부분 찾기

df[df['car_model'] =='']['car_name'].unique()
```

## 공백은 그냥 car_name에서 car_brand 뺀 데이터로 넣어주자

```python
df.fillna('',inplace = True)
none_index = df[df['car_model'] == ''].index
```

```python
for idx in none_index:
    df['car_model'][idx] = df['car_name'][idx].replace(f"{df['car_brand'][idx]} ",'')
```
